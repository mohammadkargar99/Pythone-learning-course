```python
import numpy as np
import pandas as pd 
```

```python
DF.describe()
```

```python
DF.info()
```

```python
#Surname
DF.dropna(subset = ['Surname'], inplace=True)

#Geography
DF['Geography'].fillna(DF['Geography'].mode()[0], inplace = True)

#Gender
DF.dropna(subset = ['Gender'], inplace=True)

#Tenure
DF['Tenure'].fillna(DF['Tenure'].mode()[0], inplace = True)

#Balance
DF.dropna(subset = ['Balance'], inplace=True)

#NumOfProducts
DF['NumOfProducts'].fillna(DF['NumOfProducts'].mode()[0], inplace = True)

#HasCrCard
DF['HasCrCard'].fillna(DF['HasCrCard'].mode()[0], inplace = True)

#IsActiveMember
DF['IsActiveMember'].fillna(DF['IsActiveMember'].mode()[0], inplace = True)

#EstimatedSalary
DF['EstimatedSalary'].fillna(DF['EstimatedSalary'].mean(), inplace = True)

#Exited
DF['Exited'].fillna(DF['Exited'].mode()[0], inplace = True)
```

```python
print(f'Is there any null values in dataset?? {DF.isnull().any().values.any()}')
```

## 2. Inconsistent Data:

```python
for col in DF.columns:
    print(f'{col :30s} >>>>      {DF[col].unique()}')
```

```python
#Geography
DF['Geography'].replace(to_replace=['3','4','56','Pooya','10000'], value=DF['Geography'].mode()[0], inplace=True)

#Gender
DF['Gender'].replace(to_replace=['9','5'], value=DF['Gender'].mode()[0], inplace=True)

#Age
DF['Age'].replace(to_replace=-10, value=DF['Age'].mode()[0], inplace=True)

#HasCaCard
DF['HasCrCard'].replace(to_replace=-1000, value=DF['HasCrCard'].mode()[0], inplace=True)

#IsActiveMember 
DF['IsActiveMember'].replace(to_replace=-1.0, value=DF['IsActiveMember'].mode()[0], inplace=True)
```

```python
DF.dtypes
```

```python
#Balance
index=[]
index.append(DF['Balance'][DF['Balance'] == 'sorry'].index)
index.append(DF['Balance'][DF['Balance'] == 'hi'].index)
DF.drop(index[0], axis=0, inplace=True)
DF.drop(index[1], axis=0, inplace=True)

#Surname
Inconsistent_Surname = list(DF['Surname'][DF['Surname'] == 'H?'].index)
Inconsistent_Surname.append(list(DF['Surname'][DF['Surname'] == 'He'].index))
for i in Inconsistent_Surname:
    DF.drop(i,inplace=True)
DF.head(20)
```
