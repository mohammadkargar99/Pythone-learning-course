1. Write a NumPy program to create a 5x5 array with random values and find the minimum and maximum values.

```python
import numpy as np
array5X5 = np.random.random((5,5))
print("5X5 array:\n",array5X5)
print('\n\nminimum_values = ',array5X5.min())
print('maximum_values = ',array5X5.max())
```

3. Write a NumPy program to create random vector of size 15 and replace the maximum value by -1.

```python
random_vec = np.random.random(15)
print(f'random_vec:\n{random_vec}')
random_vec[random_vec.argmax()]=-1
print('\n\nreplace the maximum value of random_vec by -1:\n',random_vec)
```

4. normalize an array so the values range exactly between 0 and 1?

```python
array = np.random.randint(0, 1000, (1, 100))
normalizedData = (array-np.min(array))/(np.max(array)-np.min(array))
print('array:\n',array)
print('\n\n\nnormalizedData:\n',normalizedData)
```

6. from 2010 to 2018 group weekdays and weekends in two lists

```python
date_time_range = np.arange(np.datetime64('2010-01-04'), np.datetime64('2018-01-01'))

t = int(date_time_range.size/7)
date_time_range_new = date_time_range.reshape((t, 7))

weekdays = list(date_time_range_new)
weekends = list(date_time_range_new[:, 6])
print('weekdays of 2010 to 2018:\n',weekdays)
print('\n\n\n\nsweekends of 2010 to 2018:\n',weekends)
```

8. Consider 2 points p0, p1 describing a line (2d) and a point p, how to compute distance from p to this line?

```python
def distance(p0, p1, p):
    T = p1 - p0
    L = (T**2).sum(axis=1)
    U = -((p0[:,0]-p[...,0])*T[:,0] + (p0[:,1]-p[...,1])*T[:,1]) / L
    U = U.reshape(len(U),1)
    D = p0 + U*T - p
    return np.sqrt((D**2).sum(axis=1))

p0 = np.random.uniform(-10, 10, (1, 2))
p1 = np.random.uniform(-10, 10, (1, 2))
p  = np.random.uniform(-10, 10, (1, 2))
print(distance(p0, p1, p))
```
